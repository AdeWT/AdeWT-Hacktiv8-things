#### 1. Getting rid of white space in text data

```python
#data cleaning step 1: strip, get rid of whitespace using .strip()

stripProd = df1[' Product '].str.strip()
stripDiscBand = df1[' Discount Band '].str.strip()
stripMonthName = df1[' Month Name '].str.strip()

df1[' Product '] = stripProd
df1[' Discount Band '] = stripDiscBand
df1[' Month Name ']= stripMonthName

print(df1[' Product '].unique())
print(df1[' Discount Band '].unique())
print(df1[' Month Name '].unique())
```

```python
#data cleaning step 1.1: renaming column to remove whitespace

df1 = df1.rename(columns=lambda i: i.strip())
df1.columns
```

#### 2. Removing any weird characters and spaces from numerical values, plus changing their type to int/float

Data cleaning step 2: cleaning and changing type of:
- Units Sold
- Manufacturing Price
- Sale Price
- Gross Sales
- Discounts
- Sales
- COGS
- Profit

```python
# for Units Sold column:
import numpy as np

df1['Units Sold'] = df1['Units Sold'].str.replace('$','') # deleting the dollar signs
df1['Units Sold'] = df1['Units Sold'].str.replace(',','') # deleting the commas
df1['Units Sold'] = df1['Units Sold'].astype(float) #turning data to float
df1['Units Sold'] = df1['Units Sold'].apply(np.floor) #rounding down because assuming you can't sell half a unit
df1['Units Sold'] = df1['Units Sold'].astype(int) # turning to integer
df1['Units Sold']
```

```python
#for Discounts column:

df1['Discounts'] = df1['Discounts'].str.replace('$','') # deleting the dollar signs
df1['Discounts'] = df1['Discounts'].str.replace(',','') # deleting the commas
df1['Discounts'] = df1['Discounts'].str.replace('-','0') # replacing '-' to 0
df1['Discounts'] = df1['Discounts'].astype(float) #converting to float
df1['Discounts']
```

```python
#for Profit:

df1['Profit'] = df1['Profit'].str.replace('$','') # deleting the dollar signs
df1['Profit'] = df1['Profit'].str.replace(',','') # deleting the commas
df1['Profit'] = df1['Profit'].str.replace(' ','') # deleting any space char
# So apparently the negatives are represented with (val) instead of -val, so:
df1['Profit'] = df1['Profit'].str.replace('(','-') #to change the ( to - to signify this is a negative value
df1['Profit'] = df1['Profit'].str.replace(')','') #removing the )

for n, i in enumerate(df1['Profit']):
     if i == '-':
         df1['Profit'][n] = 0 
     '''
     this loop is for removing the profit=0 entries where it's just '-'.
     We shouldn't delete all '-' in the column using .str.replace('-','') because negative values exist.
     Without the -, there won't be any negative values, then the data would be wrong.
     '''



df1['Profit'] = df1['Profit'].astype(float) #to change the data type to float

df1['Profit'][232] #shows a negative value to check if it is formatted correctly (stays negative)

df1['Profit'] #check the data in the column again


    
```

```python
# making loop for cleaning other data columns

toClean = ['Manufacturing Price', 'Sale Price', 'Gross Sales', 'Sales', 'COGS']
# the above comment is the list for reference. If the column name is in the list, the data cleaning process is then done.

for x in df1.columns:
    if x in toClean:
        df1[x] = df1[x].str.replace('$','') #remove the dolla sign
        df1[x] = df1[x].str.replace(',','') #remove the comma
        df1[x] = df1[x].str.replace(' ','') #remove any space char
        df1[x] = df1[x].astype(float) #turn the type from object to float

df1
```

### 3. Data Normalization

Do 3NF normalization, make sure there are 5 tables after normalization (incl. main table), pay attention to usage of primary and foreign key, and then save all tables to csv, name them with names that will be used in the database. Index of dataframe does not have to be included as table schema will apply auto-increment for primary key.

#### Plan: 
1. Have segment marked by ID in the main table, refer to segment table.
2. Have countries marked by ID in the main table, refer to country table.
3. Have product marked by ID in the main table, refer to product table.
4. Have discount band marked by ID in the main table, refer to discount band table.

```python
# step 1: get the unique values of each of those 4 columns

segment1 = df1['Segment'].unique()
country1 = df1['Country'].unique()
product1 = df1['Product'].unique()
discoBandID1 = df1['Discount Band'].unique()

# step 2: have the empty table ready to be appended

segmentTable = {'Segment ID':[], 'Segment':[]}
countryTable = {'Country ID':[], 'Country':[]}
productTable = {'Product ID':[], 'Product':[]}
discoBandTable = {'Discount Band ID':[], 'Discount Band':[]}

# step 3: use loops to populate the table with the id and its description

for i, x in enumerate(segment1):
    segmentTable['Segment ID'].append(i+1)
    segmentTable['Segment'].append(x)

for i, x in enumerate(country1):
    countryTable['Country ID'].append(i+1)
    countryTable['Country'].append(x)

for i, x in enumerate(product1):
    productTable['Product ID'].append(i+1)
    productTable['Product'].append(x)

for i, x in enumerate(discoBandID1):
    discoBandTable['Discount Band ID'].append(i+1)
    discoBandTable['Discount Band'].append(x)

#step 4: turn the tables into pandas DataFrame

df2segment = pd.DataFrame.from_dict(segmentTable)
df3country = pd.DataFrame.from_dict(countryTable)
df4product = pd.DataFrame.from_dict(productTable)
df5disco = pd.DataFrame.from_dict(discoBandTable)

# step 5: check/see the DataFrame result

print(df2segment)
print(df3country)
print(df4product)
print(df5disco)
```
