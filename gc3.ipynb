```python
# setting up empty lists to store values from webscraping
nameProd = []
priceProd = []
sellerProd = []
townProd = []
amountProd = []
ratingProd = []

# webscraping part:
for i in range(1,21):
    #the part to continue "turning the page"
    url="https://www.tokopedia.com/search?navsource=&page={}&q=seblak&srp_component_id=02.01.00.00&srp_page_id=&srp_page_title=&st=".format(i)
    driver.get(url)
    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")

    # getting the name:
    for member in soup.find_all('div', {'class':'css-uwyh54'}):
        try:
            nameProd += [nameProd.get_text() for nameProd  in member.find('div', {'class':'prd_link-product-name css-3um8ox'})]
        except:
            nameProd.append(None)

    # getting the price:
        try:
            priceProd += [priceProd.get_text() for priceProd in member.find('div', {'class':'prd_link-product-price css-h66vau'})]
        except:
            priceProd.append(None)

    # getting the seller name:
        try:
            sellerProd += [sellerProd.get_text() for sellerProd in member.find('span', {'class':'prd_link-shop-name css-1kdc32b flip'})]
        except:
            sellerProd.append(None)

    # getting the seller town:
        try:
            townProd += [townProd.get_text() for townProd in member.find('span', {'class':'prd_link-shop-loc css-1kdc32b flip'})]
        except:
            townProd.append(None)
    
    # getting the amount sold:
        try:
            amountProd += [amountProd.get_text() for amountProd in member.find('span', {'class':'prd_label-integrity css-1sgek4h'})]
        except:
            amountProd.append(None)

    # gettting rating:
        try:
            ratingProd += [ratingProd.get_text() for ratingProd in member.find('span', {'class':'prd_rating-average-text css-t70v7i'})]
        except:
            ratingProd.append(None)
    


# # checking if length all same so turning to pandas dataframe won't be a problem
# print('amount of data in column nameProd: ', len(nameProd))
# print('amount of data in column priceProd: ', len(priceProd))
# print('amount of data in column sellerProd: ', len(sellerProd))
# print('amount of data in column townProd: ', len(townProd))
# print('amount of data in column amountProd: ', len(amountProd))
# print('amount of data in column ratingProd: ', len(ratingProd))
```

```python
# #setting a DataFrame and populating the columns then seeing the data

seblakDF = pd.DataFrame()

seblakDF['Name'] = nameProd
seblakDF['Price'] = priceProd
seblakDF['Seller'] = sellerProd
seblakDF['Amount Sold'] = amountProd
seblakDF["Seller's town"] = townProd
seblakDF['Rating'] = ratingProd

seblakDF
```

#### 2. Data cleaning:

As the data will be used for exploring ratings, amount of sales, and seller's town, it's irrelevant to keep empty data, hence the rows where those columns are empty will be purged from the table.

```python
# dropping missing values

SebDF.dropna(inplace=True)

#to see amount of rows and missing values remaining:
SebDF.info()
```

Insight: 14 rows with missing values have been removed, data now consists only of data with zero null.

We can drop the index/unnamed column as it's not going to be used for our exploration / analysis.

```python
# dropping the column unnamed

SebDF.drop('Unnamed: 0', axis='columns', inplace=True)

#to see the change
SebDF.info()
```

Let's clear out the formatting of the numerical data so that it can be turned into int/float after that.

```python
# check how data is formatted:

SebDF.iloc[1:10]
```

Insight :

- Price: remove 'Rp' and dot
- Amount sold: remove '+' and the word 'terjual', turn 'rb' into 000

```python
#cleaning price column
SebDF['Price'] = SebDF['Price'].str.replace('Rp','')
SebDF['Price'] = SebDF['Price'].str.replace('.','')

#cleaning amount sold column
SebDF['Amount Sold'] = SebDF['Amount Sold'].str.replace('+','')
SebDF['Amount Sold'] = SebDF['Amount Sold'].str.replace(' terjual','')
SebDF['Amount Sold'] = SebDF['Amount Sold'].str.replace('rb', '000')

#to check the change
SebDF.iloc[1:10]
```

The column of price and amount sold can now be turned into integer (there's no comma on prices and amount sold, so there's no need for float type).

```python
#turning column price and amount into int

SebDF['Price'] = SebDF['Price'].astype(int)
SebDF['Amount Sold'] = SebDF['Amount Sold'].astype(int)

#to see the change
SebDF.info()
```
